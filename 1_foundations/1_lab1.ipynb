```python
# Primeiro vamos fazer uma importação
from dotenv import load_dotenv
```

```python
# Agora é hora de carregar as chaves da API nas variáveis de ambiente

load_dotenv(override=True)
```

```python
# Verificar as chaves

import os
openai_api_key = os.getenv('OPENAI_API_KEY')

if openai_api_key:
    print(f"Chave da API da OpenAI existe e começa com {openai_api_key[:8]}")
else:
    print("Chave da API da OpenAI não configurada - vá para o guia de solução de problemas na pasta setup")
    
```

```python
# E agora - a declaração de importação mais importante
# Se você receber um erro de importação - vá para o guia de solução de problemas

from openai import OpenAI
```

```python
# E agora vamos criar uma instância da classe OpenAI
# Se você não tem certeza do que significa criar uma instância de uma classe - vá para a pasta guides!
# Se você receber um NameError - vá para a pasta guides para aprender sobre NameErrors
# Não está criando uma Instância do ChatGPT. É uma biblioteca leve para conexão com os endpoints (pontos de extremidade) do OpenAI na nuvem.
openai = OpenAI()
```

```python
# Criar uma lista de mensagens no formato familiar do OpenAI

messages = [{"role": "user", "content": "Quanto é 2+2?"}]
```

```python
# E agora chame! Qualquer problema, vá para o guia de solução de problemas
# Isso usa GPT 5 nano, o modelo incrivelmente barato

response = openai.chat.completions.create(
    model="gpt-5-nano",
    messages=messages
)

print(response.choices[0].message.content)
```

```python
# E agora - vamos fazer uma pergunta:

question = "Por favor, proponha uma pergunta difícil e desafiadora para avaliar o QI de alguém. Responda apenas com a pergunta."
messages = [{"role": "user", "content": question}]
```

```python
# pergunte - isso usa GPT 5 mini, ainda barato mas mais poderoso que o nano

response = openai.chat.completions.create(
    model="gpt-5-mini",
    messages=messages
)

question = response.choices[0].message.content

print(question)
```

```python
# form a new messages list
messages = [{"role": "user", "content": question}]
```

```python
# Pergunte novamente

response = openai.chat.completions.create(
    model="gpt-5-mini",
    messages=messages
)

answer = response.choices[0].message.content
print(answer)
```

```python
from IPython.display import Markdown, display

display(Markdown(answer))

```

```python
# Primeiro crie as mensagens:

messages = [{"role": "user", "content": "Escolha uma área de negócio para criar um negócio. Responda apenas com o nome do negócio."}]
```

```python
# Então faça a primeira chamada:
response = openai.chat.completions.create(
    model="gpt-5-mini",
    messages=messages
)

business_idea = response.choices[0].message.content

print(business_idea)

# E repita!
```

```python
# Segundo crie as mensagens:

messages = [{"role": "user", "content": f"Identifique um problema no negócio '{business_idea}'. Um problema que precisa ser explicado explicitamente. Qual é esse problema/desafio que pode ser resolvido com uma solução ágil?"}]
```

```python
# Então faça a segunda chamada:

response = openai.chat.completions.create(
    model="gpt-5-mini",
    messages=messages
)

business_pain_point = response.choices[0].message.content

print(business_pain_point)
```

```python
from IPython.display import Markdown, display

display(Markdown(business_pain_point))
```

```python
# Terceiro crie as mensagens:

messages = [{"role": "user", "content": f"Proponha uma solução de IA para o problema a seguir: {business_pain_point}"}]
```

```python
# Então faça a terceira chamada:

response = openai.chat.completions.create(
    model="gpt-5-mini",
    messages=messages
)

business_solution_by_ai = response.choices[0].message.content

print(business_solution_by_ai)
```

```python
from IPython.display import Markdown, display

display(Markdown(business_solution_by_ai))
```
