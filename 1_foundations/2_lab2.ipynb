```python
# Comece com os imports — peça ao ChatGPT para explicar qualquer pacote que você não conheça

import os
import json
from dotenv import load_dotenv
from openai import OpenAI
from anthropic import Anthropic
from IPython.display import Markdown, display
```

```python
# Lembre-se sempre de fazer isto!
load_dotenv(override=True)
```

```python
# Imprima os prefixos das chaves para ajudar em qualquer depuração

openai_api_key = os.getenv('OPENAI_API_KEY')
anthropic_api_key = os.getenv('ANTHROPIC_API_KEY')
google_api_key = os.getenv('GOOGLE_API_KEY')
deepseek_api_key = os.getenv('DEEPSEEK_API_KEY')
groq_api_key = os.getenv('GROQ_API_KEY')

if openai_api_key:
    print(f"A chave de API da OpenAI existe e começa com {openai_api_key[:8]}")
else:
    print("Chave de API da OpenAI não definida")
    
if anthropic_api_key:
    print(f"A chave de API da Anthropic existe e começa com {anthropic_api_key[:7]}")
else:
    print("Chave de API da Anthropic não definida (e isto é opcional)")

if google_api_key:
    print(f"A chave de API do Google existe e começa com {google_api_key[:2]}")
else:
    print("Chave de API do Google não definida (e isto é opcional)")

if deepseek_api_key:
    print(f"A chave de API da DeepSeek existe e começa com {deepseek_api_key[:3]}")
else:
    print("Chave de API da DeepSeek não definida (e isto é opcional)")

if groq_api_key:
    print(f"A chave de API da Groq existe e começa com {groq_api_key[:4]}")
else:
    print("Chave de API da Groq não definida (e isto é opcional)")
```

```python
request = "Por favor, elabore uma pergunta desafiadora e sutil que eu possa fazer a diversos LLMs para avaliar sua inteligência. "
request += "Responda apenas com a pergunta, sem explicação."
messages = [{"role": "user", "content": request}]
```

```python
messages
```

```python
openai = OpenAI()
response = openai.chat.completions.create(
    model="gpt-5-mini",
    messages=messages,
)
question = response.choices[0].message.content
print(question)
```

```python
competitors = []
answers = []
messages = [{"role": "user", "content": question}]
```

```python
# A API que conhecemos bem

model_name = "gpt-5-mini"

response = openai.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content

display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
```

```python
# A Anthropic tem uma API um pouco diferente, e Max Tokens é obrigatório

model_name = "claude-3-7-sonnet-latest"

claude = Anthropic()
response = claude.messages.create(model=model_name, messages=messages, max_tokens=1000)
answer = response.content[0].text

display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
```

```python
gemini = OpenAI(api_key=google_api_key, base_url="https://generativelanguage.googleapis.com/v1beta/openai/")
model_name = "gemini-2.0-flash"

response = gemini.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content

display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
```

```python
deepseek = OpenAI(api_key=deepseek_api_key, base_url="https://api.deepseek.com/v1")
model_name = "deepseek-chat"

response = deepseek.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content

display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
```

```python
groq = OpenAI(api_key=groq_api_key, base_url="https://api.groq.com/openai/v1")
model_name = "llama-3.3-70b-versatile"

response = groq.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content

display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
```

## Para a próxima célula, usaremos o Ollama

O Ollama executa um serviço web local que fornece um endpoint compatível com OpenAI,  
e executa modelos localmente usando código C++ de alto desempenho.

Se você não tiver o Ollama, instale-o acessando https://ollama.com, clicando em Download e seguindo as instruções.

Depois de instalado, você deve conseguir acessar: http://localhost:11434 e ver a mensagem "Ollama is running"

Talvez seja necessário reiniciar o Cursor (e talvez o computador). Depois, abra um Terminal (control+\`) e execute `ollama serve`

Comandos úteis do Ollama (execute-os no terminal, ou com um ponto de exclamação neste notebook):

`ollama pull <model_name>` baixa um modelo localmente  
`ollama ls` lista todos os modelos que você baixou  
`ollama rm <model_name>` exclui o modelo especificado dos seus downloads

```python
ollama = OpenAI(base_url='http://localhost:11434/v1', api_key='ollama')
model_name = "llama3.2"

response = ollama.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content

display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
```

```python
# Então, onde estamos?

print(competitors)
print(answers)
```

```python
# É bom saber como usar "zip"
for competitor, answer in zip(competitors, answers):
    print(f"Competidor: {competitor}\n\n{answer}")
```

```python
# Vamos juntar tudo — observe o uso de "enumerate"

together = ""
for index, answer in enumerate(answers):
    together += f"# Resposta do competidor {index+1}\n\n"
    together += answer + "\n\n"
```

```python
print(together)
```

```python
judge = f"""Você está julgando uma competição entre {len(competitors)} competidores.
Cada modelo recebeu esta pergunta:

{question}

Seu trabalho é avaliar cada resposta quanto à clareza e à força do argumento e classificá-las em ordem do melhor para o pior.
Responda com JSON, e somente JSON, com o seguinte formato:
{{"results": ["número do melhor competidor", "número do segundo melhor competidor", "número do terceiro melhor competidor", ...]}}

Aqui estão as respostas de cada competidor:

{together}

Agora responda com o JSON com a ordem ranqueada dos competidores, nada mais. Não inclua formatação markdown nem blocos de código."""
```

```python
print(judge)
```

```python
judge_messages = [{"role": "user", "content": judge}]
```

```python
# Hora do julgamento!

openai = OpenAI()
response = openai.chat.completions.create(
    model="o3-mini",
    messages=judge_messages,
)
results = response.choices[0].message.content
print(results)
```

```python
# OK, vamos transformar isso em resultados!

results_dict = json.loads(results)
ranks = results_dict["results"]
for index, result in enumerate(ranks):
    competitor = competitors[int(result)-1]
    print(f"Posição {index+1}: {competitor}")
```
