```python
# As importações

from dotenv import load_dotenv
from agents import Agent, Runner, trace

```

```python
# O ponto de partida usual

load_dotenv(override=True)
```

```python

# Crie um agente com nome, instruções, modelo

agent = Agent(name="Piadista", instructions="Você é um contador de piadas", model="gpt-5-mini")
```

```python
# Execute a piada com Runner.run(agent, prompt) e então imprima final_output

with trace("Contando uma piada"):
    result = await Runner.run(agent, "Conte uma piada sobre agentes de IA autônomos")
    print(result.final_output)
```
